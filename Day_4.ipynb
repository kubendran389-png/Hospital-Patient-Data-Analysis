# **Reshape & Transpose**

```python
import numpy as np

a = np.array([1,2,3,4,5,6])
b = a.reshape(2,3)

print(b)

c= b.transpose()
print("\n",c)
```

# **Aggregation & Axis Operation**

```python
#sum() --> total

import numpy as np

a = np.array([1,2,3,4,5,6])
total = np.sum(a)
print(total)
```

```python
#mean --> average

import numpy as np

a = np.array([6,2,1,6,3,9])
avg = np.mean(a)
print(avg)
```

```python
#median() --> middle value when sorted

import numpy as np

a = np.array([6,2,1,6,3,9])
median = np.median(a)
print(median)
```
