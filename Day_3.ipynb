# **ARRAY OPERATION**

```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])

#addition
c = a+b
print(c)

#multiplication
d= a*b
print(d)

#power
e = a**2
print(e)

#sqrt
f = np.sqrt(a)
print(f)

#sum
g = np.sum(a)
print(g)
```
