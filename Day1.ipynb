```python
#grading system

def get_grade(mark):
    if mark >=90:
      return "S"
    elif mark >=75:
      return "A"
    else:
      return "F"

marks={
    "AK": 90,
    "Bala": 80,
    "Kavi": 70
}

passed = 0
failed = 0

for name, mark in marks.items():
  grade = get_grade(mark)
  print(name,":",mark,":","--Grade:",grade)

  if grade=="F":
    failed+=1
  else:
    passed+=1

print("Passsed students:",passed)
print("failed students:",failed)
```
