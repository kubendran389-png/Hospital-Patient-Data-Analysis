# **DAY 5 TASK EMPLOYEE DATASET EXPLORATION**

```python
"""DAY 5 TASK Employee dataset exploration"""

import pandas as pd
employees={
    "Emp_ID":[101,102,103,104,105],
    "Name":["Arun","Priya","Rahul","Divya","Kumar"],
    "Department":["IT","HR","IT","Finance","HR"],
    "Salary":[50000,40000,50000,55000,42000],
    "Experience":[2,3,5,6,4]
           }
df=pd.DataFrame(employees)
print("Employee Dataset")
print(df)
print("\nAverage Salary")
print(df["Salary"].mean())
print("\nHighest Salary")
print(df["Salary"].max())
print("\nEmployees in HR")
print(df[df["Department"]=="HR"])
print("\nEmployees with experience>3")
print(df[df["Experience"]>3])
print("\nEmployee count by department")
print(df["Department"].value_counts())
```

# **DAY 7 TASK Clean a messy sales dataset in python**

```python


import pandas as pd
data={
     "Product":["Laptop","Mobile","None","Tablet","Laptop"],
     "Sales":[50000,30000,20000,None,50000],
     "Quantity":[5,3,None,2,5]
}
df=pd.DataFrame(data)
print("original dataset:")
print(df)
df=df.drop_duplicates()
df["Product"]=df["Product"].fillna("Unknown")
df["Sales"]=df["Sales"].fillna(df["Sales"].mean())
df["Quantity"]=df["Quantity"].fillna(0)
print("\nCleaned Dataset:")
print(df)
```

# **Group By**

# **Day 8:-Task Region-wise sales report**

```python


import pandas as pd
data={
    "Region":["North","South","North","South","East","West"],
    "Product":["Laptop","Mouse","keyboard","Laptop","Monitor","Laptop"],
    "Sales":[10000,20000,30000,40000,50000,60000]
}
df=pd.DataFrame(data)
df_active=df[df["Sales"]>0]
df_sorted=df_active.sort_values(["Region","Sales"],ascending=[True,False])
reported=df_sorted.groupby("Region")["Sales"].agg(["sum","mean","max","count"])
print(reported)
```

# **Monthly sales using bar chart and line chart**

```python


import matplotlib.pyplot as plt
months=["Jan","Feb","Mar","Apr","May","June"]
sales=[10000,12000,15000,13000,18000,20000]
plt.figure(figsize=(18,5))
#line chart
plt.subplot(1,3,1)
plt.plot(months,
         sales,
         marker="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
#Bar Chart
plt.subplot(1,3,2)
plt.bar(months,
        sales,
        color="orange")
plt.title("Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")
#PIE CHART
plt.subplot(1,3,3)
expenses=[5000,3000,2000]
categories=["Food","Travel","Shopping"]
plt.pie(expenses,
        labels=categories,
        autopct="%1.1f%%",
        colors=["pink","yellow","green"])
plt.title("Expenses distribution")
plt.tight_layout()
plt.show()
```
